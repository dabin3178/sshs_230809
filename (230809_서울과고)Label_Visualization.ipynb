```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
wellness = pd.read_csv('/content/drive/MyDrive/LAMDA/mutil_labeled_wellness_data(kkm).csv')
wellness.head(50)
```

```python
criteria = list(wellness.columns.values)
criteria = criteria[2:]
criteria
```

```python
# 각 라벨별 개수 보기

counts = []
for criterion in criteria:
  counts.append((criterion, wellness[criterion].sum()))
df_stats = pd.DataFrame(counts, columns=['criteria ', 'number of texts'])
df_stats
```

```python
# 라벨별 개수 그래프로 나타내기

sns.set(font_scale = 1)
plt.figure(figsize=(8,4))

ax= sns.barplot(x=criteria, y=wellness.iloc[:,2:].sum().values)

plt.title("Texts in each criterion", fontsize=12)
plt.ylabel('Number of texts', fontsize=9)
plt.xlabel('Text Type ', fontsize=9)

#adding the text labels
rects = ax.patches
labels = wellness.iloc[:,2:].sum().values
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height + 5, label, ha='center', va='bottom', fontsize=9)

plt.show()
```

```python
# 라벨을 n개 가지는 텍스트의 수 계산

rowSums = wellness.iloc[:,2:].sum(axis=1)
multiLabel_counts = rowSums.value_counts()
multiLabel_counts = multiLabel_counts.iloc[1:]

sns.set(font_scale = 1)
plt.figure(figsize=(8,4))

ax = sns.barplot(x=multiLabel_counts.index, y=multiLabel_counts.values)

plt.title("Texts having multiple labels ")
plt.ylabel('Number of texts', fontsize=9)
plt.xlabel('Number of labels', fontsize=9)

#adding the text labels
rects = ax.patches
labels = multiLabel_counts.values
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height + 5, label, ha='center', va='bottom')

plt.show()
```
